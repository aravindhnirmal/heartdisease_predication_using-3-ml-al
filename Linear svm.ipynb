```python
import pandas as pd
import numpy as np

#heart = pd.read_csv("heart.csv")
heart =pd.read_csv(r'F:\sem-6\miniproone\dataset.csv')
heart.describe()
heart.shape
labels= np.array(heart.iloc[:,-1:])
print(labels)

features= np.array(heart.iloc[:,:13])
print(features)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=0)
from sklearn.preprocessing import StandardScaler
# define min max scaler
scaler = StandardScaler()
# transform data
X_train_scaled = scaler.fit_transform(X_train)

print(X_train_scaled)

X_test_scaled = scaler.fit_transform(X_test)
print(X_test_scaled)
```

# linear Kernel


```python
from sklearn.svm import SVC
svm_linear = SVC(kernel='linear', C=0.01)
svm_linear.fit(X_train_scaled, y_train)
print("Accuracy:", svm_linear.score(X_train_scaled, y_train))
```

```python
print("Accuracy:", svm_linear.score(X_test_scaled, y_test))
```

```python

import numpy as np

# create a new data point with 13 features
new_data = np.array([[59, 1, 0, 135, 234, 0, 1, 161, 0, 0.5, 1, 0, 3]])

# scale the new data using the same scaler that was used to scale the training data
new_data_scaled = scaler.transform(new_data)

# make the prediction on the new data point
predicted_label = svm_linear.predict(new_data_scaled)

# print the predicted label
print("Predicted label:", predicted_label)
```
